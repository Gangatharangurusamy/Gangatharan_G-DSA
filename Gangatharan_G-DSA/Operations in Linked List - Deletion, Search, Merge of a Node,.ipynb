# Delete Node

```python
##Time complexity = O(n)"Worst case"
##Time complexity = O(1)"Best case"

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def insertAtend(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node = None
        
    
    #delete the node at any given position
    def deleteNode(self,pos):
        if self.head is None:
            return
        temp = self.head
        for i in range(pos -   1):
            temp = temp.next
            if temp is None:
                return 
        tempptr = temp.next.next
        temp.next = None
        temp.next = tempptr
    
    
    def printLikedlist(self):
        temp = self.head
        while temp:
            print(str(temp.data)+" ", end = " ")
            temp = temp.next
            
llist = LinkedList()
llist.insertAtend(10)
llist.insertAtend(20)
llist.insertAtend(30)
llist.insertAtend(40)
llist.insertAtend(50)
llist.printLikedlist()
print()

llist.deleteNode(3)
llist.printLikedlist()
```

# Search Node


```python
##Time complexity = O(n)

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def insertAtend(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node = None
        
    
    def countNode(self):
        count = 0
        temp = self.head
        while temp:
            temp = temp.next
            count += 1
        return count
    
    #search for a given node inside the Linked List
    #If node is present return Ture otherwise return False
    def searchNode(self, nodedata):
        temp = self.head
        while temp :
            if temp.data == nodedata:
                return True
            temp = temp.next
        return False  
    
   
    def printLikedlist(self):
        temp = self.head
        while temp:
            print(str(temp.data)+" ", end = " ")
            temp = temp.next
            
llist = LinkedList()
llist.insertAtend(10)
llist.insertAtend(20)
llist.insertAtend(30)
llist.insertAtend(40)
llist.insertAtend(50)
llist.insertAtend(60)
llist.insertAtend(70)
llist.insertAtend(80)
llist.printLikedlist()
print()
res = llist.countNode()
print("Count of number of nodes in the given Linked List:", res)
print()

nodedata = 80
if llist.searchNode(nodedata):
    print("Data is present at Linked List")
else:
    print("No data is present at Linked List")
    
```

# Merge two sorted Linked List give final sorted Linked List

```python
##Time cmplixity = O(n)

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def insertAtend(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node = None
        
   
    def printLikedlist(self):
        temp = self.head
        while temp:
            print(str(temp.data)+" ", end = " ")
            temp = temp.next

            
# merge twosorted Linked List
#output : 10 -> 11 -> 20 -> 22 -> 30 -> 33 -> 40 -> 44  -> 50 -> 55
def mergeLists(head1, head2):
    temp = None
    ##llist1 is empty
    if head1 is None:
        return head2
    ##llist2 is Empty
    if head2 is None:
        return head1
    
    if head1.data <= head2.data:
        temp = head1
        temp.next = mergeLists(head1.next, head2)
    else:
        temp = head2
        temp.next = mergeLists(head1, head2.next)
    
    return temp
    
    
    
llist1 = LinkedList()
llist2 = LinkedList()
llist1.insertAtend(10)
llist1.insertAtend(20)
llist1.insertAtend(30)
llist1.insertAtend(40)
print("Linked List-1:")
llist1.printLikedlist()
print()

llist2.insertAtend(11)
llist2.insertAtend(22)
llist2.insertAtend(33)
llist2.insertAtend(39)

print("Linked List-2")
llist2.printLikedlist()
print()

print("Merged Linked List")
llist3 = LinkedList()
llist3.head = mergeLists(llist1.head, llist2.head)
llist3.printLikedlist()
```
