# Reversal of linked list

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insertAtend(self,new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        
        
    def reverseNode(self):
        prev = None
        next = None
        curr = self.head
        while curr:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        self.head = prev
        
    def print(self):
        temp = self.head
        while temp:
            print(str(temp.data)+" ", end = " ")
            temp = temp.next

# Create an instance of the LinkedList class
llist = LinkedList()

# Insert elements into the linked list
llist.insertAtend(2)
llist.insertAtend(3)
llist.insertAtend(4)
llist.insertAtend(5)
llist.insertAtend(6)
print("Original")
# Print the linked list
llist.print()
print()

#Print the ReverseLIst

print("Reverseal Linked List:")
llist.reverseNode()
llist.print()
```

# Circular Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insertAtend(self,new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        
    def circularLLConversion(self):
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = self.head 
    
    def print(self):
        temp = self.head
        while temp:
            print(str(temp.data)+" ", end = " ")
            temp = temp.next

# Create an instance of the LinkedList class
llist = LinkedList()

# Insert elements into the linked list
llist.insertAtend(2)
llist.insertAtend(3)
llist.insertAtend(4)
llist.insertAtend(5)
llist.insertAtend(6)
print("Original")
# Print the linked list
llist.print()
print()
        
llist.circularLLConversion()
print("Circular Linked List:")
llist.print()
print()
```

# Palindrome Linked List

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def isPalindrome(self, head):
        # Convert the input list to a linked list
        nodes = [ListNode(val=x) for x in head]
        for i in range(len(nodes) - 1):
            nodes[i].next = nodes[i + 1]

        # Find the middle of the linked list
        fast = slow = nodes[0]
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next

        # Reverse the second half of the linked list
        prev = None
        while slow:
            temp = slow.next
            slow.next = prev
            prev = slow
            slow = temp

        # Check if it is a palindrome
        left, right = nodes[0], prev
        while right:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next
        return True

# Driver Code
s = Solution()
head = [1, 2, 2, 3, 1]
if s.isPalindrome(head):
    print("It is a Palindrome")
else:
    print("It is not a Palindrome")
```

# Linked List Cycle

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insertAtend(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

        
    ## To detect the loop inside the linked list
    def detectloop(self):
        hare = self.head
        tortoise = self.head
        while tortoise and hare and hare.next:
            hare = hare.next.next
            tortoise = tortoise.next
            if tortoise == hare:
                return True
        return False
            
    def printlinkedlist(self):
        temp = self.head
        while temp:
            print(str(temp.data) + " ", end=" ")
            temp = temp.next

# Create an instance of the LinkedList class
llist = LinkedList()

# Insert elements into the linked list
llist.insertAtend(2)
llist.insertAtend(3)
llist.insertAtend(4)
llist.insertAtend(5)
#llist.head.next.next = llist.head # for Check
if llist.detectloop():
    print("Loop is inside the Linked List ")
else:
    print("No Loop")
# Print the linked list

```
